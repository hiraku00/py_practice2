## Part 4: scikit-learnを使ったSVMの実際

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.style.use('seaborn-whitegrid')
#sns.set_style('whitegrid')
```

アヤメのデータをimportするところから始めます。

```python
from sklearn import datasets
```

```python
iris = datasets.load_iris()
```

```python
X = iris.data
Y = iris.target
print(iris.DESCR)
```

パラメータについて、ドキュメントを参照してみてください。

[SVC](http://scikit-learn.org/stable/modules/generated/sklearn.svm.SVC.html) (Support Vector Classification) 

[SVM library of Sci Kit Learn](http://scikit-learn.org/stable/modules/svm.html)


```python
from sklearn.svm import SVC
```

```python
# インスタンスを生成
model = SVC()
```

```python
# データを分ける
# train_test_split : https://docs.pyq.jp/python/machine_learning/tips/train_test_split.html
from sklearn.cross_validation import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, random_state=3)
```

```python
# トレーニング
model.fit(X_train, Y_train)
```

```python
# 精度
from sklearn import metrics
```

```python
pred = model.predict(X_test)
```

```python
print(metrics.accuracy_score(Y_test, pred))
```

非常に高い予測精度が得られました。

デフォルトでは、RBFカーネルが使われています。

それぞれのカーネルの違いを[scikit-learnのドキュメント](http://scikit-learn.org/stable/modules/svm.html#svm-classification)に詳しく載っています。

これを自分で作る方法を書いておきますので、興味がある方はやってみてください。

```python
from sklearn import svm
```

```python
# 図示できるのが2次元までなので、変数を2つに絞る
X = iris.data[:,:2]
Y = iris.target
```

```python
# SVMの正則化パラメータ
C = 1.0
```

```python
# SVC with a Linear Kernel
svc = svm.SVC(kernel='linear', C=C).fit(X, Y)
```

```python
# Gaussian Radial Basis Function
rbf_svc = svm.SVC(kernel='rbf', gamma=0.7, C=C).fit(X, Y)
```

```python
# SVC with 3rd degree poynomial
poly_svc = svm.SVC(kernel='poly', degree=3, C=C).fit(X, Y)
```

```python
# SVC Linear
lin_svc = svm.LinearSVC(C=C).fit(X, Y)
```

```python
# step size
h = 0.02

# X軸の最大最小
x_min = X[:, 0].min() - 1
x_max = X[:, 0].max() + 1

# Y軸の最大最小
y_min = X[:, 1].min() - 1
y_max = X[:, 1].max() + 1
```

```python
# meshgridを作成
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
```

```python
titles = ['SVC with linear kernel',
          'LinearSVC (linear kernel)',
          'SVC with RBF kernel',
          'SVC with polynomial (degree 3) kernel']
```

```python
for i, clf in enumerate((svc, lin_svc, rbf_svc, poly_svc)):
    # 境界線を描画
    plt.figure(figsize=(15, 15))
    plt.subplot(2, 2, i+1)
    
    plt.subplots_adjust(wspace=0.4, hspace=0.4)
    
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    
    Z = Z.reshape(xx.shape)
    
    plt.contourf(xx, yy, Z, cmap=plt.cm.terrain, alpha=0.5, linewidths=0)
    
    plt.scatter(X[:, 0], X[:, 1], c=Y, cmap=plt.cm.Dark2)
    
    plt.xlabel('Sepal Length')
    plt.ylabel('Sepal Width')
    plt.xlim(xx.min(), xx.max())
    plt.ylim(yy.min(), yy.max())
    plt.xticks(())
    plt.yticks(())
    plt.title(titles[i])
```
